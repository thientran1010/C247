## Activity 3

Present the user with a list of 7-9 items (such as the list created in the previous activities) and prompt them to enter one item to delete from the list. Delete the named item from the list and display the updated list.

```python
# your code here
print(data)
print("Enter 1 item that you would like to remove from the list: ")
tbr = input()
if tbr in data:
  data.remove(tbr)
  print("Item {} removed successfully".format(tbr))
else:
  print("item not on the list")
```

```python
# your code here
print(data)
print("Enter one item on the list to be updated")
tbd=input()
print("Enter the value you want to change into")
newVal=input()
if tbd in data:
  data[data.index(tbd)] = newVal
  print("Item updated")
else:
  print("Item not found")
print(data)
```
