## Practice 5b
The tuple below contains the abbreviation of all the states in the USA. 

Use slicing to create three new tuples: 
* One tuple that contains all states that start with N
* One tuple that contains all states that start with O
* One tuple that contains all states that start with A

```python
states = ("AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DE", "FL", "GA", 
          "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
          "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY")

# your code here 
def findInd(target,states):
  start =-1 
  end = -1
  
  for i in range(len(states)):
    if(states[i][0] == target and start == -1):
      start=i
    if(states[i][0]!= target and start != -1 and end == -1):
      end = i
  return [start,end]
l = findInd('A',states)
a = states[l[0]:l[1]]
l = findInd('O',states)
o = states[l[0]:l[1]]
l = findInd('N',states)
n = states[l[0]:l[1]]
print(a,o,n)
  
```
